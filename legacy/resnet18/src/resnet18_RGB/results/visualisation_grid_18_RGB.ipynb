```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as mticker
import seaborn as sns
```

```python
#import des résultats

grid_classique = pd.read_csv("/autofs/unityaccount/cremi/leanguye/projet-deep-learning-m1/resnet18/src/resnet18_RGB/results/grid_search_results.csv")
grid_par_classe = pd.read_csv("/autofs/unityaccount/cremi/leanguye/projet-deep-learning-m1/resnet18/src/resnet18_RGB/results/grid_search_results_par_classe.csv")
grid_par_classe_aug = pd.read_csv("/autofs/unityaccount/cremi/leanguye/projet-deep-learning-m1/resnet18/src/resnet18_RGB/results/grid_search_results_par_classe_data_augmentation.csv")
```

```python
#pour sélectionner les configs avec 50 epochs / 100 epochs
grid_classique_50 = grid_classique.head(32)
grid_par_classe_50 = grid_par_classe.head(32)
grid_par_classe_aug_50 = grid_par_classe_aug.head(32)

grid_classique_100 = grid_classique.tail(32)
grid_par_classe_100 = grid_par_classe.tail(32)
grid_par_classe_aug_100 = grid_par_classe_aug.tail(32)
```

```python
#extraite les configs

config_classique_50 = grid_classique_50['config'].str.extract(r'(config\d+)')[0].str.replace("config", "c")
config_classique_100 = grid_classique_100['config'].str.extract(r'(config\d+)')[0].str.replace("config", "c")

config_par_classe_50 = grid_par_classe_50['config'].str.extract(r'(config\d+)') [0].str.replace("config", "c")
config_par_classe_100 = grid_par_classe_100['config'].str.extract(r'(config\d+)') [0].str.replace("config", "c")

config_par_classe_aug_50 = grid_par_classe_aug_50['config'].str.extract(r'(config\d+)') [0].str.replace("config", "c")
config_par_classe_aug_100 = grid_par_classe_aug_100['config'].str.extract(r'(config\d+)') [0].str.replace("config", "c")


#autre manière de faire : labels = [f"c{i}" for i in range(len(df_random))] mais ne permet pas de s'assurer que c'est la bonne config

```

```python
# Création de la colonne learning_rate, gamma, et step
grid_par_classe_aug_50['learning_rate'] = (grid_par_classe_aug_50['config'].str.extract(r'lr([0-9.]+)').astype(float))
grid_par_classe_aug_50['step'] = (grid_par_classe_aug_50['config'].str.extract(r'step([0-9]+)').astype(int))
grid_par_classe_aug_50['gamma'] = (grid_par_classe_aug_50['config'].str.extract(r'gamma([0-9.]+)').astype(float))
```
