```python
import matplotlib.pyplot as plt
from collections import Counter

from src.data import load_data_train

x_train, y_train = load_data_train()
counts = Counter(y_train) #compte le nombre d'occurence des labels

labels = ["Health", "Other", "Rust"]
values = [counts[0], counts[1], counts[2]] #counts[0] = nombre de 0 dans y_train

plt.figure(figsize=(6,4))
plt.bar(labels, values)
plt.title("Distribution des images par classe")
plt.xlabel("Classes")
plt.ylabel("Nombre d'images")
plt.show() #on a 200 images par classes
```
