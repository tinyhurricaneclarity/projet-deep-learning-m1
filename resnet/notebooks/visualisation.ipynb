## Visualisation de la distribution des classes

```python
import matplotlib.pyplot as plt
from collections import Counter
import sys #permet d’accéder aux paramètres internes de Python, notamment sys.path (liste des dossiers)
import os #permet de manipuler des fichiers
sys.path.append(os.path.abspath("..")) #cherche dans le dossier parent (src est pas dans le dossier notebooks)
from src.data import load_data_train


path_train_rgb = "/net/cremi/leanguye/projet-deep-learning-m1/resnet/data/beyond-visible-spectrum-ai-for-agriculture-2026/Kaggle_Prepared/train/RGB/"
path_train_val = "/net/cremi/leanguye/projet-deep-learning-m1/resnet/data/beyond-visible-spectrum-ai-for-agriculture-2026/Kaggle_Prepared/train/val/"
x_train, y_train = load_data_train(path_train_rgb, path_train_val)

counts = Counter(y_train) #compte le nombre d'occurence des labels

labels = ["Health", "Other", "Rust"]
values = [counts[0], counts[1], counts[2]] #counts[0] = nombre de 0 dans y_train

plt.figure(figsize=(6,4))
plt.bar(labels, values)
plt.title("Distribution des images par classe")
plt.xlabel("Classes")
plt.ylabel("Nombre d'images")
plt.show() #on a 200 images par classes
```
